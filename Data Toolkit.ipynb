# 1 What is NumPy, and why is it widely used in Python?

  -  NumPy, short for Numerical Python, is a fundamental library for numerical and scientific computing in Python. It provides support for multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.
  - It's widely used because it provides:
  - Speed: Vectorized operations are implemented in C under the hood, making
   NumPy much faster than vanilla Python for numerical tasks.
  - Memory Efficiency: Arrays take up less space compared to Python lists.
  - Foundation for Other Libraries: Libraries like Pandas, SciPy, TensorFlow, scikit-learn, and PyTorch use NumPy arrays as their base.
  - Community & Documentation: It's mature, well-documented, and widely supported.

# 2 How does broadcasting work in NumPy?

  - Broadcasting in NumPy allows arrays of different shapes to be used together in arithmetic operations. It works by automatically expanding the smaller array along the missing dimensions so that both arrays have the same shape, without copying data.

# 3 What is a Pandas DataFrame?

  - A Pandas DataFrame is a two-dimensional, labeled data structure in Python, similar to a table in a database or an Excel spreadsheet.

# 4  Explain the use of the groupby() method in Pandas?

  - The groupby() method in Pandas is used to split data into groups, apply operations, and then combine the results. It's very powerful for aggregating, transforming, or filtering data based on some criteria.
  - Aggregation (sum(), mean(), count(), etc.)
  - Transformations (transform(), apply())
  - Filtering (filter() based on group stats)

# 5 Why is Seaborn preferred for statistical visualizations?

  - Seaborn is preferred for statistical visualizations in Python because it is built specifically for making complex statistical plots easy and beautiful. It’s built on top of Matplotlib and tightly integrated with Pandas.

# 6 What are the differences between NumPy arrays and Python lists?

  - NumPy arrays are faster, use less memory, and support vectorized operations. They are homogeneous (same data type) and great for numerical/matrix operations.
  - Python lists are slower, use more memory, and are heterogeneous (mixed data types). They don’t support element-wise math directly.

# 7 What is a heatmap, and when should it be used?

  - A heatmap is a data visualization that uses color to represent values in a matrix or 2D dataset.
  - When to Use a Heatmap:
  - To show correlations between variables.
  - To visualize a confusion matrix in classification tasks.
  - To explore patterns or trends in large data tables.
  - To represent missing data, intensity, or frequency visually.

# 8 What does the term “vectorized operation” mean in NumPy?

  - Vectorized operation in NumPy means performing operations on entire arrays (vectors) without using explicit loops.

# 9 How does Matplotlib differ from Plotly?

  - Matplotlib and Plotly are both popular Python libraries for data visualization, but they have key differences:
  - Matplotlib: Primarily for static plots, good for detailed, publication-quality visualizations but requires more customization.
  - Plotly: Best for interactive plots, easy to use, and provides beautiful, dynamic visualizations for web applications

# 10 What is the significance of hierarchical indexing in Pandas?

  - Hierarchical indexing in Pandas (also called MultiIndex) allows you to work with multiple levels of indexing on rows and columns. It’s useful when you need to handle complex, multi-dimensional data in a 2D table
  Significance of hierarchical indexing in Pandas:
  - Multi-level organization: Group data based on multiple keys (e.g., time series data with year and month).
  - Efficient data access: Allows you to access data using multiple levels of labels (e.g., accessing data by both "region" and "product").
  - Flexible slicing: You can slice, group, and aggregate data efficiently across multiple levels.

# 11 What is the role of Seaborn’s pairplot() function?

  - Seaborn's pairplot() creates a grid of scatter plots for all pairs of numerical columns in a dataset, along with diagonal histograms to show distributions. It's useful for exploring relationships and patterns between variables.

# 12 What is the purpose of the describe() function in Pandas?

  - The describe() function in Pandas provides summary statistics of a DataFrame or Series, including measures like:

  - Count: Number of non-null entries
  - Mean: Average value
  - Std: Standard deviation
  - Min/Max: Minimum and maximum values
  - 25%, 50%, 75%: Percentiles (quartiles)

# 13 Why is handling missing data important in Pandas?

  - Handling missing data is important in Pandas because missing or NaN (Not a Number) values can lead to:

  - Incorrect Analysis: Operations like mean, sum, or group-by can be skewed or fail when missing values are present.

  - Data Integrity: Missing values can affect the quality and reliability of the results.

  - Model Accuracy: In machine learning, missing data can distort models and lead to incorrect predictions if not handled properly.

# 14 What are the benefits of using Plotly for data visualization?

  - Plotly offers interactive, visually appealing plots, is easy to use, and supports a wide range of visualizations like 3D plots and maps. It's great for web integration and building interactive dashboards.

# 15  How does NumPy handle multidimensional arrays?

  - NumPy handles multidimensional arrays using the ndarray (N-dimensional array) object, which can represent arrays of any number of dimensions.

# 16 What is the role of Bokeh in data visualization?

  - Bokeh is a Python library for creating interactive, web-based visualizations. It supports a wide range of plot types, real-time data streaming, and extensive customization for professional visualizations.

# 17 Explain the difference between apply() and map() in Pandas?

  - apply(): Works on both Series and DataFrames, applies a function along rows/columns, and supports complex operations.
  - map(): Works only on Series, applies a function element-wise, and is typically used for simple transformations like mapping values.

# 18  What are some advanced features of NumPy?

  - Broadcasting: Allows operations on arrays of different shapes.

  - Vectorization: Enables element-wise operations without loops, improving performance.

  - Fancy Indexing: Indexing arrays with other arrays or lists to access specific elements.

  - Linear Algebra: Built-in functions for matrix operations like dot product, eigenvalues, and matrix decomposition.

  - Random Sampling: Provides tools for generating random numbers and distributions.

  - Memory Views: Efficient handling of large datasets by creating views of arrays without copying data.

# 19  How does Pandas simplify time series analysis?

  - Pandas simplifies time series analysis by offering easy date handling, resampling, rolling statistics, and tools for shifting and missing data handling.

# 20 What is the role of a pivot table in Pandas?

  - A pivot table in Pandas is used to summarize and aggregate data, making it easier to analyze and explore. It reorganizes data by reshaping it, allowing you to:

  - Group data based on one or more columns.
  - Aggregate values (e.g., sum, mean, count) for the grouped data.
  - Display results in a tabular format with multi-level indexing.

# 21 Why is NumPy’s array slicing faster than Python’s list slicing?

  - NumPy's array slicing is faster because it works with contiguous memory, low-level C optimizations, and views (instead of copies), making it more efficient than Python's list slicing.

```python
# 4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 2, 3, 2],
    'D': [5, 6, 7, 8, 9]
}
df = pd.DataFrame(data)
corr = df.corr()
sns.heatmap(corr, annot = True)
plt.show()
```

```python
# 5 Generate a bar plot using Plotly

import plotly.express as px
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 15, 7, 12]
}
df = pd.DataFrame(data)
fig = px.bar(df, x='Category', y='Values', title='Bar Plot')
fig.show()
```

```python
# 8 Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]     # Line 1: y = x^2
y2 = [0, 1, 2, 3, 4, 5]       # Line 2: y = x
y3 = [0, -1, -2, -3, -4, -5]

plt.plot(x , y1 , c = "red")
plt.plot(x , y2 , c = "blue")
plt.plot(x , y3 , c = "green")
plt.legend()
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
